```python
import geopandas as gpd
import pandas as pd
```

# **Data cleanining for County**

```python
county = 'county.geojson'
const = 'constituency.geojson'

county = gpd.read_file(county)
const = gpd.read_file(const)

county = county.drop('Name_code', axis = 1)
```

```python
# remove / and white space and replace with - 
county['Name'] = county['Name'].str.replace('/| ', '-')
county['Name'] = county['Name'].replace('Murang?a', 'Murang\'a')
county['Name'] = county['Name'].str.upper()

county.columns = ['name','code','geometry']

county
```

```python
county.to_file('countys.geojson', driver='GeoJSON')
```

# **Data cleanining for Constituency**

```python
const.head()
```

```python
const.columns = ['county','const_code','const','county_code','geometry']
const['const_code'] = const['const_code'].astype(int)
const['county_code'] = const['county_code'].astype(int)
const['county'] = const['county'].str.replace(' ', '-')
const['county'] = const['county'].replace('THARAKA--NITHI', 'THARAKA-NITHI')
col = const.pop('county_code')
const.insert(0,col.name, col)
const
```

```python
const.to_file('constituencys.geojson', driver='GeoJSON')
```
